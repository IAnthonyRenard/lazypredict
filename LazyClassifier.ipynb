```python
from lazypredict.Supervised import LazyClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from tabulate import tabulate

def load_data():
    data = load_breast_cancer()
    X = data.data
    y = data.target
    return X, y

def train_test_split_data(X, y, test_size=0.5, random_state=123):
    return train_test_split(X, y, test_size=test_size, random_state=random_state)

def display_results(models):
    print(tabulate(models, headers='keys', tablefmt='pretty'))

def main():
    # Chargement des donn√©es
    X, y = load_data()

    # Division des donn√©es en ensembles d'entra√Ænement et de test
    X_train, X_test, y_train, y_test = train_test_split_data(X, y)

    # Initialisation du mod√®le LazyClassifier
    clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)

    # Entra√Ænement du mod√®le et obtention des pr√©dictions
    models, predictions = clf.fit(X_train, X_test, y_train, y_test)

    # Affichage des r√©sultats sous forme de tableau
    display_results(models)

if __name__ == "__main__":
    main()
```
